```python
#1. 캠퍼스픽>공모전>IT/소프트웨어/게임 링크를 main_url로 두고 게시되어 있는 각 공모전의 상세 링크를 sub_url에 저장합니다.

main_url = "https://www.campuspick.com/contest?category=108"
base_url = "https://www.campuspick.com"

# Chrome 웹 드라이버 설치 및 초기화
driver = webdriver.Chrome(options=chrome_options)
driver.get(main_url)

# 페이지가 로드될 때까지 대기
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.ID, 'container')))

sub_url = []  # 순서를 유지하기 위해 list 사용
last_height = driver.execute_script("return document.body.scrollHeight")

while len(sub_url) < 760:
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    new_links = soup.select("#container > div.list > div > a")

    for link in new_links:

        absolute_url = urljoin(base_url, link.get('href'))

        # base_url과 동일하지 않고, sub_url에 없으면 추가
        if absolute_url != base_url and absolute_url not in sub_url:
            sub_url.append(absolute_url)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

driver.quit()

print(f"Total links collected: {len(sub_url)}")
for url in sub_url:
    print(url)
```

```python
# 리스트를 2개로 나누는 함수
def split_list(urls, n):
    # 각 부분 리스트의 크기를 계산
    k, m = divmod(len(urls), n)
    return [urls[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n)]

# 리스트를 4개로 나누기
split_urls = split_list(sub_url, 2)

# 각 리스트에 이름 부여
list1, list2 = split_urls
```

```python
# 2. 크롤링 함수 만들기
# 1) 각자 맡은 list에 저장된 웹페이지 해당 주소로 이동하여 크롤링 함수 실행
# 2) 크롤링 함수
# 공모전 상세 페이지에서 공모전명, 주최기관명, 기간, 상금, 공모전 소개를 가져옵니다.
# 요약 칼럼의 경우 공모전 소개의 내용을 30% 요약하는 코드의 반환값입니다.
# 위의 데이터를 엑셀에 계속 추가하는 함수를 만듭니다.
```

```python
def scrape_contest_info(sub_url):
    results = []
    driver = webdriver.Chrome(options=chrome_options)  # 웹드라이버 초기화

    for i in range(len(sub_url)):
        driver.get(sub_url[i])
        time.sleep(3)
        wait = WebDriverWait(driver, 20)

        # 공모전명
        try:
            c_name = driver.find_element(By.TAG_NAME, 'h1').text
        except:
            c_name = "없음"

        # 태그

        #'IT/소프트웨어/게임'은 공통적으로 있으므로 제거하는 함수
        def remove_after_char(s, char):
            result = []
            for c in s:
                if c == char:
                    break  # 특정 글자를 만나면 반복 종료
                result.append(c)  # 그렇지 않으면 글자를 저장

            # 리스트를 문자열로 변환해서 반환
            return ''.join(result)

        try:
          badges = driver.find_element(By.CSS_SELECTOR, "p.badges").text
          badge_result = remove_after_char(badges, "I")
        except:
          badges = "없음"


        # 주최기관명
        try:
            c_company = driver.find_element(By.CSS_SELECTOR, "p.company").text
        except:
            c_company = "없음"

        # 기간
        try:
            due_date = driver.find_elements(By.CSS_SELECTOR, "p.indent")[0].text
        except:
            due_date = "없음"

        # 상금
        try:
            prize_money = 0
            p_element = driver.find_elements(By.CSS_SELECTOR, "p.indent")

            # p_element에서 텍스트를 추출
            text_content = [element.text for element in p_element]

            # 모든 텍스트를 하나의 문자열로 결합
            combined_text = " ".join(text_content)

            # 정규식으로 금액 패턴 찾기
            match = re.search(r"\d{1,3}(,\d{3})*원", combined_text)

            # 금액이 발견되었을 때 출력
            if match:
                prize_money = match.group(0)
        except:
            prize_money = "0"

        #공모 주제
        try:
            # article.description의 텍스트를 줄 단위로 가져오기
            c_subject = driver.find_element(By.CSS_SELECTOR, 'article.description')
            all_text = c_subject.text.split('\n')

            # \n이 나올 때까지의 문장 저장
            capture = False
            captured_text = ""

            for line in all_text:
                if capture:
                    if line == "":  # 빈 줄이 나오면 종료
                        break
                    captured_text += line + " "
                if '공모 주제' in line:  # '공모주제'가 포함된 줄 찾기
                    capture = True
        except:
            captured_text = "없음"

        # 공모전 소개
        try:
            description = driver.find_element(By.CSS_SELECTOR, "article.description").text.split('\n')
        except:
            description = "없음"

        # 요약 정보
        try:
            description = driver.find_element(By.CSS_SELECTOR, "article.description").text
            summary = summarizer.summarize(description, ratio=0.3)  # 원문의 약 30%만을 요약하여 반환
        except:
            #description = "없음"
            summary = '없음'

        # 결과에 정보 저장
        result = {'공모전명': c_name, '태그': badge_result, '기관명': c_company, '기간': due_date, '상금': prize_money, '주제': captured_text, '공모전 소개': description, '요약': summary}
        results.append(result)

    driver.quit()  # 웹드라이버 종료
    return pd.DataFrame(results)
```

```python
# 2) 날짜 형식 통일하기: 예. 2022-01-01
from datetime import datetime

#년도가 없는 공모전은 임의로 현재 년도를 넣습니다.
current_year = datetime.now().year

# 날짜 문자열을 변환하는 함수
def convert_date(date_str):
    # 요일 정보 제거
    date_str = date_str.split('(')[0].strip()

    # 년, 월, 일을 분리
    date_parts = date_str.split()
    if len(date_parts) == 3:
        year, month, day = date_parts
        year = int(year[:-1])  # '년' 제거
    else:
        year = current_year
        month, day = date_parts

    month = int(month[:-1])  # '월' 제거
    day = int(day[:-1])  # '일' 제거

    # 날짜 문자열을 datetime 객체로 변환
    return datetime(year, month, day).strftime('%Y-%m-%d')

df['시작 날짜'] = df['시작 날짜'].apply(convert_date)
df['종료 날짜'] = df['종료 날짜'].apply(convert_date)

df
```

```python
# 1) 필터링 함수 정의
# 매개변수: df, tag, 기관종류, 난이도
# 기본값: None
# 각 매개변수에 입력한 값이 None이 아닌 경우, 각 컬럼의 값이 입력값과 일치하는 행만 선택하도록 함수 구성

def filter_contests(df, tag_input, 기관종류_input, 난이도_input):
    # 원본을 바꾸지 않기 위해 df 복사
    filtered_df = df.copy()

    # tag 필터링
    if tag_input is not None:
        filtered_df = filtered_df[filtered_df['태그'] == tag_input]

    # 기관종류 필터링
    if 기관종류_input is not None:
        filtered_df = filtered_df[filtered_df['기관종류'] == 기관종류_input]

    # 난이도 필터링
    if 난이도_input is not None:
        filtered_df = filtered_df[filtered_df['난이도'] == 난이도_input]

    # 상금에 따라 내림차순으로 정렬
    filtered_df = filtered_df.sort_values(by='상금', ascending=False)

    return filtered_df
```

```python
import pandas as pd

def clean_data(df):
    # 모든 열을 문자열로 변환하고 공백 제거
    df = df.applymap(lambda x: ' '.join(str(i).strip() for i in x) if isinstance(x, list) else str(x).strip() if pd.notnull(x) else x)
    return df



def filter_contests(df, 태그, 기관종류, 난이도):
    # 데이터 클리닝
    df = clean_data(df)

    # 입력값에 대한 공모전 종류 필터링 (부분 매칭 허용)
    if 태그:
        df = df[df['태그'].str.contains(태그, case=False, na=False)]
        print(f"'{태그}'에 해당하는 공모전: {len(df)}개")  # 깔끔한 디버깅용 출력

    # 입력값에 대한 기관 종류 필터링
    if 기관종류:
        df = df[df['기관종류'].str.strip() == 기관종류]
        print(f"'{기관종류}'에 해당하는 공모전: {len(df)}개")  # 깔끔한 디버깅용 출력

    # 입력값에 대한 난이도 필터링
    if 난이도:
        df = df[df['난이도'].str.strip() == 난이도]
        print(f"'{난이도}' 난이도에 해당하는 공모전: {len(df)}개")  # 깔끔한 디버깅용 출력

    return df

# 입력 받기
tag_input = input("공모전 종류를 선택하세요 (기획, 아이디어, 사진, 영상, UCC, 디자인, 예체능, 문학, 수기, 시나리오, 브랜드, 네이밍, 광고, 마케팅): ").strip()
기관종류_input = input("기관 종류를 선택하세요 (공공기관/사기업): ").strip()
난이도_input = input("난이도를 선택하세요 (상/중/하): ").strip()

# 필터링된 결과 출력
result_df = filter_contests(df, tag_input, 기관종류_input, 난이도_input)

# 조건에 맞는 결과가 없을 경우 메시지 출력
if result_df.empty:
    print("조건에 맞는 공모전이 없습니다.")
else:
    # 상위 10개 데이터 출력
    print(result_df.head(10))

# 전체 결과 반환
result_df


```

```python
# 2) 추천 함수 생성
# 매개변수: df, contest_name, 공모전 간 코사인 유사도 행렬
def recommendations(df, contest_name, cosine_sim):
    # 입력한 공모전 이름과 일치하는 '공모전명' 컬럼 내 인덱스를 idx에 할당
    idx = df[df['공모전명'] == contest_name].index[0]

    # 해당 공모전과 모든 공모전 간의 유사도 점수를 가져오기
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도 점수를 기준으로 공모전 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 3개의 공모전 점수 가져오기
    sim_scores = sim_scores[1:4]  # 첫 번째는 자기 자신이므로 제외

    # 공모전 인덱스 가져오기
    contest_index = [i[0] for i in sim_scores]

    # 공모전 인덱스를 이용하여 '공모전명' 칼럼에서 상위 3개의 유사한 공모전을 찾고 유사도 점수 출력
    recommended_contests = df['공모전명'].iloc[contest_index]
    similarity_scores = [i[1] for i in sim_scores]

    # enumerate, range 등을 사용하여 추천 공모전 출력
    # 출력 예: 1. 2021 OCR 인공지능 학습데이터 해커톤 | Similarity Score: 0.41
    for i, (contest, score) in enumerate(zip(recommended_contests, similarity_scores)):
        print(f"{i+1}. {contest} | Similarity Score: {score:.2f}")
```

```python
def recommend_contests(df, contest_name, cosine_sim):
    # 주어진 contest_name에 해당하는 공모전의 index를 반환
    contest_index = df[df['공모전명'] == contest_name].index[0]

    # 주어진 공모전의 코사인 유사도 행렬에서 최근접 이웃 indices를 호출
    sim_scores = list(enumerate(cosine_sim[contest_index]))

    # 유사도 점수를 기준으로 공모전 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 자기 자신을 제외한 상위 3개의 공모전 인덱스 호출
    top_contests = sim_scores[1:4]  # 첫 번째는 자기 자신이므로 제외

    # 최근접 이웃의 공모전명을 호출
    recommended_contests = df.iloc[[i[0] for i in top_contests]]

    # 추천된 공모전의 유사도 점수와 함께 반환
    recommendations = [(row.공모전명, score) for row, score in zip(recommended_contests.itertuples(), [i[1] for i in top_contests])]

    # 추천 결과 출력 (enumerate 사용)
    for i, (contest, score) in enumerate(recommendations):
        print(f"{i+1}. {contest} | Similarity Score: {score:.2f}")
```
